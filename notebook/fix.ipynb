```python
import pandas as pd
import os

# Go one level up → then into data folder
raw_df = pd.read_csv("../data/database.csv")

# Clean dataset is already in notebook folder
clean_df = pd.read_csv("cleaned_vehicle_recall_data.csv")
```

```python
raw_df["Recall Notification Date"] = raw_df["Recall Notification Date"].fillna(0)
raw_df["Recall Notification Date"] = raw_df["Recall Notification Date"].astype(int).astype(str)

raw_df["Recall Notification Date"] = pd.to_datetime(
    raw_df["Recall Notification Date"],
    format="%Y%m%d",
    errors="coerce"
)
```

```python
raw_df["Recall Notification Date"].head(10)
```

```python
raw_df["Recall Year"] = raw_df["Recall Notification Date"].dt.year
```

```python
raw_df["Recall Year"].value_counts().head()
```

```python
clean_df["Recall Notification Date"] = raw_df["Recall Notification Date"]
clean_df["Recall Year"] = raw_df["Recall Year"]
```

```python
clean_df.to_csv("cleaned_vehicle_recall_data_updated.csv", index=False)
```
